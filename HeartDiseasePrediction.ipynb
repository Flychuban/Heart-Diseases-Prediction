```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
```

```python
data = pd.read_csv('/content/drive/MyDrive/HeartDiseasePrediction/data/heart_disease_data.csv')
```

```python
# Extracting labels from data

Y = data['target']
X = data.drop(['target'], axis=1)
```

```python
# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2, stratify=Y, random_state=2)
```

```python
X_train
```

```python
regressor = LogisticRegression()
```

```python
# Train the model
regressor.fit(X_train, Y_train)
```

Evaluate the model

```python
y_pred = regressor.predict(X_test)
acc = accuracy_score(Y_test, y_pred)
```

```python
print(f"Accuracy: {acc}")
```

```python
# Save the model
filename = "heart_disease_model.sav"
pickle.dump(regressor, open(filename, 'wb'))
```

```python
del regressor
```

```python
regressor = pickle.load(open(filename, 'rb'))
```

```python
# Make prediction system

# 0 -> NO heart disease
# 1 -> Heart disease

input_data = (63,1,3,145,233,1,0,150,0,2.3,0,0,1) # This is example of heart disease
input_data = np.asarray(input_data) # Making it to array
input_data = np.expand_dims(input_data, axis=0) # This helps for prediction cuz it wants a batch of data not single input

prediction = regressor.predict(input_data)
if prediction[0] == 0:
  print("This person hasn't heart disease!")
else:
  print("This person HAS heart disease!")
```
